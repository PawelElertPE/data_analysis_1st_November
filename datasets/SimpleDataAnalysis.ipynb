```python
filtered_data = insurance_data[insurance_data['age'] > 45]

filtered_data.head()
```

```python
filtered_data = insurance_data[(insurance_data['age'] > 45) & (insurance_data['smoker'] == 'yes')]

filtered_data.head()
```

```python
filtered_data = insurance_data[(insurance_data['bmi'] > 30) | (insurance_data['smoker'] == 'yes')]

filtered_data.head()
```

```python
trimmed_data = insurance_data.drop(columns=['children'])

trimmed_data.head()
```

```python
average_charges_smokers = insurance_data[insurance_data['smoker'] == 'yes']['charges'].mean()

average_charges_non_smokers = insurance_data[insurance_data['smoker'] == 'no']['charges'].mean()
```

```python
print(f'Average charges for smokers: {average_charges_smokers:.2f}')
print(f'Average charges for non-smokers: {average_charges_non_smokers:.2f}')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.scatter(insurance_data['bmi'], insurance_data['charges'], alpha=0.5)
plt.title('BMI vs. Insurance Charges')
plt.xlabel('BMI')
plt.ylabel('Insurance Charges')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))  # Adjust the figure size as needed

# Create a box plot
sns.boxplot(x='smoker', y='charges', data=insurance_data)
plt.title('Insurance Charges by Smoking Status (Box Plot)')
plt.xlabel('Smoker')
plt.ylabel('Insurance Charges')
plt.show()
```
